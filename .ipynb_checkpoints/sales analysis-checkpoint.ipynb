```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
#reading data
df = pd.read_csv(".\Sales_Data\Sales_April_2019.csv")
```

```python
#combining data from different files into one
data = [file for file in os.listdir('.\Sales_Data')]
all_data = pd.DataFrame()
for file in data:
    df = pd.read_csv("./Sales_Data/" + file)
    all_data=pd.concat([all_data,df])
    
all_data.to_csv("all_data.csv", index=False)
```

```python
#loading concatenated data
new = pd.read_csv("all_data.csv")
new.head()
```

```python
#checking if the NaN are necessary and dropping them if not
new[new.isna().any(axis=1)]
#545 rows of NaN to be removed


new_dat = new.dropna()

#Checking if na is removed
new_dat[new_dat.isna().any(axis=1)]

new_dat = new_dat[new_dat['Order Date'].str[0:2] !='Or']
```

```python
#Adding a month Column 
new_dat['Month']=new_dat['Order Date'].str[0:2]
new_dat['Month']=new_dat['Month'].astype('int32')
new_dat.head()
```

```python
#converting colums to their respective data type

new_dat['Quantity Ordered'] = pd.to_numeric(new_dat['Quantity Ordered'])
new_dat['Price Each'] = pd.to_numeric(new_dat['Price Each'])
```

```python
#Adding a sales column

new_dat['Sales']= new_dat['Quantity Ordered']* new_dat['Price Each']
```

```python
new_dat.head()
results = new_dat.groupby('Month').sum()
```

```python
#plotting the monthly sales to better understnad the sales distribution
months = range(1,13)
plt.bar(months, results['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD($)')
plt.xlabel('Month')
plt.show()
```

```python
#creating a column for the location. Using .apply() to split at the location of ,

def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]


new_dat['City']= new_dat['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")
new_dat.head()
```

```python
#plotting the location wise sales
results_loc = new_dat.groupby('City').sum()
results_loc
```

```python
import matplotlib.pyplot as plt
cities = [city for city, df in new_dat.groupby('City')]

plt.bar(cities, results_loc['Sales'])
plt.xticks(cities,rotation = 'vertical', size = 7)
plt.ylabel('Sales in USD($)')
plt.xlabel('City name')
plt.show()
```

```python
#converting order date to date time to find out most sales timing
new_dat['Order Date']= pd.to_datetime(new_dat['Order Date'])



new_dat['Hour']=new_dat['Order Date'].dt.hour
new_dat.head()
```

```python
#plotting data by hours to see the peak hours
hours = [hour for hour, df in new_dat.groupby('Hour')]

plt.plot(hours, new_dat.groupby(['Hour']).count())
plt.xticks(hours)
plt.xlabel('Hours of the day')
plt.ylabel('Orders')
plt.grid()
plt.show()
```

```python
#checking order ID to see what products where sold together

datf = new_dat[new_dat['Order ID'].duplicated(keep = False)]
datf['Grouped'] = datf.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
datf = datf[['Order ID','Grouped']].drop_duplicates()

datf.head(20)

#Above data helps make business decisions
```

```python
#checking the sales per product

prod_group = new_dat.groupby('Product')
quant_ord = prod_group.sum()['Quantity Ordered']


prod = [prod for prod, df in prod_group]

plt.bar(prod, quant_ord) 
plt.xticks(prod,rotation = 'vertical',size=7)
plt.xlabel('Product')
plt.ylabel('Quantity Ordered')
plt.show()
```

```python
#plotting price against the items sold
prices = new_dat.groupby('Product').mean()['Price Each']

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.bar(prod, quant_ord, color='g')
ax2.plot(prod,prices,'b-')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price($)',color ='blue')
ax1.set_xticklabels(prod, rotation='vertical', size =7)
plt.show()
```
